```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%ls
```

```python
s_df = pd.read_excel('Steel4340_raw_data.xls')
s_df.head()
```

```python
a_df = pd.read_csv('Raw Data T533.csv')
a_df.head()
```

```python
d = 0.506
r = d/2
A = np.pi*r**2

stress_a = (a_df['FORCE']/A)*0.001
strain_a = a_df['CH5']*0.01

stress_s = (s_df['FORCE']/A)*0.001
strain_s = s_df['CH5']*0.01
```

```python
plt.plot(strain_s, stress_s,strain_a, stress_a)
plt.legend(['steel','aluminum'])
plt.show()
```

```python
import os
os.listdir(os.path.join(os.getcwd(),'data'))
```

```python
d = 0.506
r = d/2
A = np.pi*r**2
fig,ax = plt.subplots(figsize=(10,10))
legend_lst =[]
for f in os.listdir(os.path.join(os.getcwd(),'data'))[3:5]:
    df = pd.read_excel(os.path.join(os.getcwd(),'data',f))
    try:
        stress = (df['FORCE']/A)*0.001
        strain = df['CH5']*0.01
        
        ax.plot(strain, stress)
        legend_lst.append(f)
    except:
        pass
ax.legend(legend_lst)
plt.show()
        
```
